Next, we want to import the data

```python
cupolaWifiData = pd.read_csv("../data/wifi_data.csv", low_memory=False)
```

```python
df = pd.DataFrame(cupolaWifiData)
df
```

Realistically, we only care about the amount of people at a certain date, so we'll make a new table with only those two values

```python
wifiData_dateAmount = df[['Date Time', 'Total Online Devices']]
wifiData_dateAmount
```

Ok, now let's check if there's any missing values

```python
wifiData_dateAmount.isnull().sum()
```

Quite a few, let's look at them

```python
null_mask = wifiData_dateAmount.isnull().any(axis=1)
null_columns = wifiData_dateAmount[null_mask]
null_columns
```

Should note these down to ask Mark about them. But for now, let's remove these rows (also I'm shortening the variable name for comfort's sake)

```python
wifiData = wifiData_dateAmount.dropna()
wifiData.isna().sum()
```

Next, let's make the amount of visitors actually a number by removing the '#' and parsing it

```python
wifiData['Total Online Devices'] = wifiData['Total Online Devices'].str.replace('#','').astype(int)
```

It gave a warning, but the command was still correctly executed

```python
print(wifiData.dtypes)
wifiData
```

As the Date Time column features an hour range per row, it is not possible to parse it to DateTime. Instead, we will be parsing only the date portion as DateTime and use that going forward

```python
dt = wifiData['Date Time'].str.split(' ').str[0]
devices = wifiData['Total Online Devices']
```

Plotting 6 months worth of data is too clustered to be readable, so we can look at the dataset monthly.

```python
#Ignore final entry because it's an average and a special row:
wifiData['Date Time'] = pd.to_datetime(wifiData['Date Time'][:-1].str.split(' ').str[0], dayfirst=True)
```

Again, we can ignore the error. Same with the next cell

```python
wifiData['year'] = wifiData['Date Time'].dt.year
wifiData['month'] = wifiData['Date Time'].dt.month
wifiData_monthly = wifiData.groupby(['year', 'month'])
```

Now all we have to do is run a loop that iterates through each month, and we can visualize the wi-fi data!

```python
for (year, month), data in wifiData_monthly:
    plt.scatter(data['Date Time'][:-1], data['Total Online Devices'][:-1], label=month)
    plt.title(f"Online Devices in {cd.month_name[int(month)]} {int(year)}")
    plt.xlabel("Day")
    plt.ylabel("Online Devices")

    #To make it more readable, we're formatting the x axis to only show days
    date_format = DateFormatter('%d')
    plt.gca().xaxis.set_major_formatter(date_format)
    
    plt.show()
```

Curiously, there is a similar pattern in each month. It follows somewhat of a Sin curve, having distinct highs and lows, but a very defined bottom averaging at around 50 at all times. Cool!
